```python
import os
try:
  file='example.txt'
  def write_file(input_data):
    f=open(file,'w')
    f.write(input_data)
    f.close()
  def read_file():
    f=open(file,'r')
    data=f.read()
    return data
  def replace_content(data,old_word,new_word):
    l=list(data.split(" "))
    for index, i in enumerate(l):
        if i==old_word:
            l[index]=new_word

    data1=str(' '.join(l))
    f=open(file,'w')
    f.write(data1)
    f.close()
    return data1

  input_data=input("Enter data to be written in file:")
  write_file(input_data)
  data=read_file()
  old_word=input("Enter the word to be replaced:")
  new_word=input("Enter the new word to be written: ")

  new_data=replace_content(data,old_word,new_word)
  print(new_data)
except Exception as e:
  print(e)
```

###Abstraction Example

```python
class test:
  def __init__(self,a,b,c,d):
    self.__a=a
    self._b=b
    self.c=c
    self.d=d
  def test_custom1(self,v):
    return v-self.__a
  def test_custom2(self,v):
    return v+self._b
  def test_custom3(self,v):
    return v*self.c
  def __str__(self):
    return "This is my test code for abstraction"
```

```python
o=test(4,5,6,7)
o.test_custom1(7)  
```

In the above lines of code we can observe the concepts of Abstraction. The variable 'a' is declared as private variable, variable 'b' is a protected variable and variable 'c' is a public variable. Hence we can not access the variable 'a' from outside the class. We could access the variable 'b' from outside the class but in a protected form. We could easily access the public varibale 'c' from outside the class. We can also access the public variable from outside the package.

However we can use all these variables while calling the class functions from outside the class.

### Multiple Inheritance Example

Multiple inheritance is said to be implement when derive class inherits properties of more than one base class. In below example derived class inherits properties of base class 1 and base class 2.

```python
class base1:
  def funcbase1(self):
    print("Function a of base1 is executed")

class base2:
  def funcbase2(self):
    print("Function a of base2 is executed")

class derived(base1,base2):
  
  def funcderive(self):
    print("function of deriveclass is executed")
  

m=derived()
m.funcbase1()
m.funcbase2()
m.funcderive()


```

###Decorators in Python

Decorators in python is an important feature. It allows to wrap another function around the called function to add some more feature to the called function without modifying it.

```python
try:
  
  def fun_decorator(func):
    def inner_fun(x,y):
      print("Begining of inner")
      if y==0:
        print("Illegal divisible by 0")
        return
      print("End of inner")
      return func(x,y)
      
    return inner_fun

  @fun_decorator
  def div_twonums(a,b):
    return a/b

  a=int(input("Enter a:"))
  b=int(input("Enter b:"))

  print("Division of two numbers is",div_twonums(a,b))
except Exception as e:
  print(e)
```
