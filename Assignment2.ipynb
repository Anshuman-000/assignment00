#**11. In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?**

**Answer-**

rane(10) refers to range upto which the loop will run the iterations. Here in this case the no of iterations will be 10. 

range(0,10) refers to starting point and end point of iterations. Here the starting point is 0 and end point will be 9. In python the end point is excluded in a range.

range(0,10,1) refers to starting point of range is 0, end point of range is 9 and the third argument is step size. step size refers to no of steps jumped while going from present iteration to next iteration in a loop.
